# Aquecimento de SQL — Dashboard Bitcoin

Nossas perguntas são:
* Qual é o **último preço** do Bitcoin?
* Qual foi o **maior preço** já registrado e **quando**?
* Qual foi o **menor preço** já registrado e **quando**?

### LAST PRICE — Último valor coletado

Qual é o último preço do Bitcoin registrado?

```python
%sql
SELECT
valor_brl        AS last_price,
timestamp        AS last_timestamp
FROM pipe_api_bitcoin.bitcoin_data.bitcoin_data
WHERE criptomoeda = 'BTC'
ORDER BY timestamp DESC
LIMIT 1
```
